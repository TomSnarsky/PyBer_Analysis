```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["New Orleans", "Milwaukee", "Omaha", "Pittsburgh", "Toledo"]
bars_in_cities = [8.6, 8.5, 8.3, 7.9, 7.2]
x_axis = np.arange(len(bars_in_cities))
```

```python
# Create a bar chart based upon the above data

plt.bar(cities, bars_in_cities, color="blue")
```

```python
# Create the ticks for our bar chart's y axis

plt.yticks(np.arange(0,10, step=1))
```

```python
# Set the limits of the x axis
```

```python
# Set the limits of the y axis

plt.ylim(0, 9)
```

```python
# Give the chart a title, x label, and y label

plt.title("Density of Bars in Cities", fontsize=18)
plt.xlabel("Cities", fontsize=14)
plt.ylabel("Bars per 10,000 Households", fontsize=14)
```
